## 2-2 데이터 전 처리
#### 목적 : 기존의 머신러닝의 문제점을 해결하기 위해 데이터 전처리 과정을 사용한다33.png

#### 기존 머신러닝의 문제점
> 길이가 25cm 이고 무게가 150g인 도미가 있는데 우리의 모델은 빙어로 예측함!!

일단 왜 이런 문제가 발생했는지 알아보자!!  

그 전에 Numpy를 이용해서 데이터를 준비하자


```python
import numpy as np

fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
# numpy의 column_stack() 함수를 이용하면 전달받은 리수트를 일렬로 세운 다음 차례대로 연결할 수 있다!!
# 이때 연결할 list는 python의 tuple 형식으로 전달해야 한다  -> tuple 형식 : (자료) vs list 형식 : [자료]

np.column_stack(([1,2,3], [4,5,6]))
```

```python
# 이 방식을 이용해서 fish_length 와 fish_weight 를 합쳐보자
# 이전 방식보다 조금 더 세련된 방법!!

# 이전 방법
old_fish_data = [[l,w] for l, w in zip(fish_length, fish_weight)]
print(old_fish_data)

# 새로운 방법
fish_data = np.column_stack((fish_length, fish_weight))
print(fish_data[:5])
```

```python
# 비슷한 방법으로 target data를 만들어보자!
# 하지만 이전 방법과는 다르게 조금 더 세련된 방식으로 만들어볼까?

# 이전 방법
old_target_data = [1] * 35 + [0] * 15
print(old_target_data)

# 새로운 방법
# np.ones() 와 np.zeros(), np.concatenate()를 사용한다!!
fish_target = np.concatenate((np.ones(35), np.zeros(14)))
print(fish_target)
```

#### cf) np.column_stack() 과 np.concatenate()의 차이![33.png](55.png)

```python
# 이전까지는 numpy 배열의 인덱스를 섞어서 -> train set과 test set을 구성함
# 하지만 이런 방법은 조금 번거롭다!!!

# 따라서, 사이킷런에 있는 train_test_split() 함수를 사용해서 조금 더 세련되게 섞을 수 있다
# train_test_split()에 원하는 리스트나 배열을 전달하면 사용할 수 있다!
# random_state는 앞에서 사용한 np.random.seed()와 같은 역할을 한다

from sklearn.model_selection import train_test_split
train_input , test_input, train_target, test_target = train_test_split(fish_data, fish_target, random_state = 42)
```

44.png

```python
# train_input 과 test_input의 분리된 비율을 확인해보기

print(train_input.shape, test_input.shape)

print(train_target.shape, test_target.shape)

# 데이터가 잘 섞였나 확인
# 원래 도미와 빙어의 비율은 35:14 -> 2.5:1 비율인데 지금 test_target에서는 3.3:1의 비율을 보이고 있음
# 약간의 샘플링 편향이 나타났는데 이러한 문제도 해결해보자!!

print(test_target)
```

```python
# 위와 같이 샘플링 편향이 발생하는 경유 stratify 매개변수에 target data를 전달하면 클래스의 비율에 맞게 알아서 잘 나눠준다!!

train_input , test_input, train_target, test_target = train_test_split(fish_data, fish_target, stratify = fish_target, random_state = 42)

# 이런 경우 test set의 비율이 2.25:1이 되어서 이전보다 샘플링 편향이 준 것을 확인할 수 있음!!
print(test_target)
```

여기까지 모든 데이터를 준비했다!!  
이제 다시 k-최근접 이웃 모델을 훈련시켜보자

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

문제가 발생했던 25cm 이고 150g인 생선을 도미로 판단해야 하는데...  
놀랍게도 빙어로 판단한다!!

```python
print(kn.predict([[25,150]]))
```

이 생선의 위치가 어디쯤인지 직접 확인해보기 위해 산점도를 그려보자!

```python
import matplotlib.pyplot as plt

plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25,150, marker = '^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

위의 그래프로 확인하면 삼각형의 샘플은 분명이 빙어보다 도미에 가깝다고 보인다!!  
근데 왜 우리 모델은 빙어에 가깝다고 판단했을까?  
  
k-최근접 이웃은 주변의 샘플 중에서 다수인 클래스를 예측값으로 알려준다  
따라서, 삼각형의 샘플들과 가장 가까운 이웃을 찾아주는 kneighbors() 메소드를 이용해서 어떤 샘플과 근접한지 알아보자

```python
distances, indexes = kn.kneighbors([[25,150]])

print(distances)
print(indexes)
```

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25,150, marker = '^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(train_input[indexes])
print(train_target[indexes])
```

위의 그래프와 결과에서 알 수 있듯이 삼각형 샘플은 확실히 빙어와 가까이 위치한다고 판단할 수 있다  
하지만 이상한 점이 있는데...  
그래프에서는 도미의 데이터와 더 가까이 보이지 않는가?  
distances의 값과 산점도를 자세히 들여다 보면 그 답이 보인다!

산점도를 자세히 보면 x축과 y축의 범위가 다른 것을 확인할 수 있다!!  
이렇게 축의 범위가 달라서 직관적으로 보기에 도미의 데이터와 가까이 보였던 것!!  

이런 문제를 해결하기 위해서 각 축의 범위를 맞춰주는 과정이 필요하다

```python
plt.scatter(train_input[:,0], train_input[:,1])
plt.scatter(25,150, marker = '^')
plt.scatter(train_input[indexes,0], train_input[indexes,1], marker='D')
plt.xlim((0,1000)) # <- 이 코드를 추가해서 x축의 범위를 y 축의 범위와 맞춰준다
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

이렇게 각 특성의 scale을 맞춰주게 되니까 산점도가 달라진다  
이런 경우 x축의 값은 거의 영향을 미치지 않고 y축의 값이 더 큰 영향을 끼치게 될 것이다  
  
**특성 간 스케일이 다른 경우는 매우 흔하므로 이를 맞춰주는 작업이 꼭 필요하다**
  
이렇게 특성값을 일정한 기준으로 맞추는 과정을 **데이터 전처리**라고 한다  
이때 가장 흔히 사용하는 전처리 방법으로 **표준 점수**가 있다  
표준 점수는 각 특성값이 0에서 표준편차의 몇 배 만큼 떨어져 있는지를 나타낸다  
이를 통새 실제 특성값의 크기와 상관없이 동일한 조건으로 비교할 수 있다.  


```python
# train_input은 (36,2)의 행렬이다
# axis = 0으로 설정하면 각 열마다 통계값을 계산할 수 있다

print(train_input)

mean = np.mean(train_input, axis = 0)
std = np.std(train_input, axis = 0)

print(mean, std)
```

55.png

```python
# 원래 train set을 평균과 표준편차를 이용해서 정규화!

train_scaled = (train_input - mean)/ std

print(train_scaled)
```

```python
# 전처리된 데이터를 기반으로 다시 산점도 그려보기

plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(25,150, marker = '^')
plt.xlabel ('length')
plt.ylabel('weight')
plt.show()
```

왜 우리가 판단하고자 하는 삼각형 샘플만 데이터들과 동떨어져서 위치할까?  
-> 데이터는 전처리를 하고 우리가 판단하고자 하는 샘플에는 전처리를 거치지 않았기 때문!!  
  
따라서, 전처리 과정을 거칠 때는 사용하는 모든 데이터에 대해 전처리 과정을 진행시켜줘야한다

```python
new = ([25,150] - mean) / std
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0],new[1], marker = '^')
plt.xlabel ('length')
plt.ylabel('weight')
plt.show()
```

드디어 두 특성이 서로 비슷한 범위의 값을 가지게 된 것!!

```python
# 이제 이러한 데이터를 가지고 다시 모델을 훈련시켜보자!!

kn.fit(train_scaled, train_target)
```

train set만 전처리 과정을 거쳤기 때문에 우리는 test set 에서도 전처리 과정을 진행시켜야 한다

```python
test_scaled = (test_input - mean) / std
```

```python
kn.score(test_scaled, test_target)
```

```python
print(kn.predict([new]))
```

```python
distances, indexes = kn.kneighbors([new])
plt.scatter(train_scaled[:,0], train_scaled[:,1])
plt.scatter(new[0],new[1], marker = '^')
plt.scatter(train_scaled[indexes, 0 ], train_scaled[indexes, 1])
plt.xlabel ('length')
plt.ylabel('weight')
plt.show()
```
